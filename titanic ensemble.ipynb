```python
import joblib
from sklearn.model_selection import GridSearchCV
```

```python
tr_features=pd.read_csv("../train_features.csv")
tr_labels=pd.read_csv("../train_labels.csv")
```

```python
def print_results(results):
    print("BEST Params {}\n".format(results.best_params_))
    means=results.cv_results_["mean_test_score"]
    stds=results.cv_results_["std_test_score"]
    for mean,std,params in zip(means,stds,results.cv_results_["params"]):
        print("{}(+/-{}) for {}".format(round(mean,3),round(std*2,3),params))
```

```python
gb=GradientBoostingClassifier()
parameters={
    "n_estimators":[5,50,250,500],
    "max_depth":[1,3,5,7,9],
    "learning_rate":[0.01,0.1,1,10,100]
}
cv=GridSearchCV(gb,parameters,cv=5)
cv.fit(tr_features,tr_labels.values.ravel())
print_results(cv)
```

```python
cv.best_estimator_
```

```python
joblib.dump(cv.best_estimator_,"../GBmodels.pkl")
```

```python
from sklearn.ensemble import RandomForestClassifier
RandomForestClassifier().get_params()
```

```python
rf=RandomForestClassifier()
parameters={
    "n_estimators":[5,50,250,500],
    "max_depth":[4,8,16,32,None]
    }
cv=GridSearchCV(rf,parameters,cv=5)
cv.fit(tr_features,tr_labels.values.ravel())
print_results(cv)
```

```python
cv.best_estimator_
```

```python
joblib.dump(cv.best_estimator_,"../RFmodels.pkl")
```

```python
from sklearn.ensemble import StackingClassifier
estimators=[("gb",GradientBoostingClassifier()),("rf",RandomForestClassifier())]
StackingClassifier(estimators=estimators).get_params()
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
estimators=[("gb",GradientBoostingClassifier()),("rf",RandomForestClassifier())]
sc=StackingClassifier(estimators=estimators)
sc.get_params()
```

```python
parameters={
    'gb__n_estimators':[50,100],
    'rf__n_estimators':[50,100],
    "final_estimator":[LogisticRegression(C=0.1),
                       LogisticRegression(C=1),
                       LogisticRegression(C=10)],
    "passthrough":[True,False]
}
cv=GridSearchCV(sc,parameters,cv=5)
cv.fit(tr_features,tr_labels.values.ravel())
print_results(cv)
```

```python
cv.best_estimator_
```

```python
joblib.dump(cv.best_estimator_,"../SCmodels.pkl")
```

```python
from sklearn.metrics import accuracy_score,precision_score,recall_score
val_features=pd.read_csv("../val_features.csv")
val_labels=pd.read_csv("../val_labels.csv")
te_features=pd.read_csv("../test_features.csv")
te_labels=pd.read_csv("../test_labels.csv")
```

```python
gb_mdl=joblib.load("../GBmodels.pkl")
rf_mdl=joblib.load("../RFmodels.pkl")
stacked_mdl=joblib.load("../SCmodels.pkl")
```

```python
from time import time
```

```python
def evaluation_model(model,features,labels):
    start=time()
    preds=model.predict(features)
    end=time()
    accuracy=round(accuracy_score(labels,preds),3)
    precision=round(precision_score(labels,preds),3)
    recall=round(recall_score(labels,preds),3)
    print("{}--Accuracy :{}/precision :{}/recall :{}/Latency:{}ms".format(str(model).split('(')[0],
                                                                                   accuracy,
                                                                                   precision,
                                                                                   recall,
                                                                                   round((end - start)*1000, 1)))
```

```python
for mdl in[gb_mdl,rf_mdl,stacked_mdl]:
    evaluation_model(mdl,val_features,val_labels)
```

```python
evaluation_model(rf_mdl,te_features,te_labels)
```
